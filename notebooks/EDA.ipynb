```python
import seaborn as sn
```

```python
import matplotlib.pyplot as pl
```

```python
pl.figure(figsize=(15,4))
pl.subplot(1,2,1)
data["sex"].value_counts().plot(kind="pie",autopct="%.2f%%",
                                    )
pl.subplot(1,2,2)
a=sn.countplot(data=data,x="sex",color="k")
for i in a.containers:
  a.bar_label(i)
pl.show()
```

```python
pl.figure(figsize=(15,4))
pl.subplot(1,2,1)
data["smoker"].value_counts().plot(kind="pie",autopct="%.2f%%",
                                    )
pl.subplot(1,2,2)
a=sn.countplot(data=data,x="smoker",color="g")
for i in a.containers:
  a.bar_label(i)
pl.show()
```

```python
pl.figure(figsize=(15,4))
pl.subplot(1,2,1)
data["region"].value_counts().plot(kind="pie",autopct="%.2f%%",
                                    )
pl.subplot(1,2,2)
a=sn.countplot(data=data,x="region",color="b")
for i in a.containers:
  a.bar_label(i)
pl.show()
```

```python
pl.figure(figsize=(15,4))
pl.subplot(1,2,1)
data["children"].value_counts().plot(kind="pie",autopct="%.2f%%",
                                   explode=[0,0,0,.2,.1,.5] )
pl.subplot(1,2,2)
a=sn.countplot(data=data,x="children",color="r")
for i in a.containers:
  a.bar_label(i)
pl.show()
```

```python
pl.figure(figsize=(20,6))
sn.regplot(x='bmi', y='expenses', data = data,color='g')
pl.figure(figsize=(20,6))
sn.regplot(x='bmi', y='expenses', data = data,scatter=False,ci=None,order=2,color='r')
pl.show()
```

expenses is increasing with bmi 

```python
pl.figure(figsize=(20,6))
sn.kdeplot(data=data,x='expenses',hue='sex')
pl.show()
```

```python
pl.figure(figsize=(20,4))
sn.kdeplot(data=data,x='expenses',hue='smoker')
pl.show()
```

expenses is very high for smoker

```python
pl.figure(figsize=(15,4))
sn.lmplot(x='bmi', y='expenses', data = data,hue='sex',col='sex',palette='husl')
pl.figure(figsize=(20,6))
sn.lmplot(x='bmi', y='expenses', data = data,scatter=False,ci=None,order=2,hue='sex',col='sex',palette='husl')
pl.show()
```

where sex=male there after 42 bmi, expenses have droped
